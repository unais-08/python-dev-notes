```python
import matplotlib.pyplot as plt
```

```python
students = ["Aman", "Divya", "Ishaan", "Riya", "Kabir"]

ml_hours = [3, 5, 7, 9, 10]
ml_marks = [58, 68, 80, 92, 95]

cd_hours = [2, 4, 6, 8, 10]
cd_marks = [52, 60, 74, 86, 93]

math_hours = [1, 3, 5, 7, 9]
math_marks = [35, 48, 58, 70, 80]
```

```python
# Create a 1x3 subplot
fig, axs = plt.subplots(1, 3, figsize=(16, 5), constrained_layout=True)
fig.suptitle("Performance Analysis of Students", fontsize=16, fontweight="bold")

# --- Subplot 1: Machine Learning ---
axs[0].plot(ml_hours, ml_marks, marker="o", label="Machine Learning", linewidth=2)
axs[0].set_title("Machine Learning")
axs[0].set_xlabel("Hours Studied")
axs[0].set_ylabel("Marks Obtained")
axs[0].grid(True, linestyle="--", alpha=0.6)
axs[0].set_xlim(0, 11)
axs[0].set_ylim(30, 100)

for i in range(len(students)):
    axs[0].annotate(
        students[i],
        (ml_hours[i], ml_marks[i]),
        textcoords="offset points",
        xytext=(0, 8),
        ha="center",
        fontsize=9,
    )

# --- Subplot 2: Compiler Design ---
axs[1].plot(
    cd_hours, cd_marks, marker="s", label="Compiler Design", linewidth=2, color="orange"
)
axs[1].set_title("Compiler Design")
axs[1].set_xlabel("Hours Studied")
axs[1].grid(True, linestyle="--", alpha=0.6)
axs[1].set_xlim(0, 11)
axs[1].set_ylim(30, 100)

for i in range(len(students)):
    axs[1].annotate(
        students[i],
        (cd_hours[i], cd_marks[i]),
        textcoords="offset points",
        xytext=(0, 8),
        ha="center",
        fontsize=9,
    )

# --- Subplot 3: Mathematics III ---
axs[2].plot(
    math_hours,
    math_marks,
    marker="^",
    label="Mathematics III",
    linewidth=2,
    color="green",
)
axs[2].set_title("Mathematics III")
axs[2].set_xlabel("Hours Studied")
axs[2].grid(True, linestyle="--", alpha=0.6)
axs[2].set_xlim(0, 11)
axs[2].set_ylim(30, 100)

for i in range(len(students)):
    axs[2].annotate(
        students[i],
        (math_hours[i], math_marks[i]),
        textcoords="offset points",
        xytext=(0, 8),
        ha="center",
        fontsize=9,
    )

# Optional: add individual legends
for ax in axs:
    ax.legend(loc="lower right")

# Show the plot
plt.show()
```
