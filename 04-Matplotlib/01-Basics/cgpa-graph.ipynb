```python
# students = ["Aman", "Divya", "Ishaan", "Riya", "Kabir"]
import matplotlib.pyplot as plt

sgpa_us = [8.94, 6.26, 6.29, 6.57, 7.50, 7.39, 8.34, 10]

cgpa_us = [8.9, 7.57, 7.07, 6.93, 7.30, 7.25, 7.97, 8.97]

sgpa_ash = [8.50, 6.90, 6.50, 6.90, 7.50, 7.44, 8.20, 10]

cgpa_ash = [8.50, 7.80, 7.20, 7.10, 7.30, 7.25, 7.90, 8.89]

sem = ["1st", "2nd", "3rd", "4th", "5th", "6th", "7th", "8th"]
# SGPA and CGPA for Us
plt.plot(sem, sgpa_us, marker='o', label="Us - SGPA")
plt.plot(sem, cgpa_us, marker='s', label="Us - CGPA")

# SGPA and CGPA for Ash
plt.plot(sem, sgpa_ash, marker='^', label="Ash - SGPA")
plt.plot(sem, cgpa_ash, marker='D', label="Ash - CGPA")

# Titles and labels
plt.title("Semester-wise SGPA and CGPA Comparison")
plt.xlabel("Semester")
plt.ylabel("GPA (out of 10)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc="lower right")

plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(13,5))

axes[0].plot(sem,sgpa_us,marker='o',label="SGPA")
axes[0].plot(sem,cgpa_us,marker='s',label="CGPA")
axes[0].set_title("Semester wise performance Us")
axes[0].set_xlabel('Semester')
axes[0].set_ylabel('SGPA(out of 10)')
axes[0].legend(loc="lower right")
axes[0].grid(True,linestyle='--',alpha=0.6)


axes[1].plot(sem,sgpa_ash,marker='o',label="SGPA")
axes[1].plot(sem,cgpa_ash,marker='s',label="CGPA")
axes[1].set_title("Semester wise performance Ash")
axes[1].set_xlabel('Semester')
axes[1].set_ylabel('SGPA(out of 10)')
axes[1].legend(loc="lower right")
axes[1].grid(True,linestyle='--',alpha=0.61)

plt.tight_layout()
plt.savefig("./figures/sem_performance.png",dpi=300, bbox_inches="tight")
plt.show()
```
