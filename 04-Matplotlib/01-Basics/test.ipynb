```python
import matplotlib.pyplot as plt
```

```python
years = [2010, 2012, 2014, 2016, 2018, 2020, 2022]
users_millions = [500, 1000, 1600, 2500, 3600, 4500, 5000]
```

```python
fig, axes = plt.subplots()

axes.plot(users_millions, years, marker="o")

axes.set_title("Global Internet Users Over Years")
axes.set_xlabel("Year", color="blue")
axes.set_ylabel("Users (in millions)")

axes.xaxis.set_tick_params(rotation=30)
axes.grid(True, linestyle="--", alpha=0.6)



plt.tight_layout()

plt.savefig("./figures/test.png", dpi=300)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator, AutoMinorLocator

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)

# Set major and minor locators for the x-axis
ax.xaxis.set_major_locator(MultipleLocator(2))  # Major tick every 2 units
ax.xaxis.set_minor_locator(AutoMinorLocator(4)) # 4 minor ticks between major ticks

# Customize major and minor tick parameters
ax.tick_params(which='major', length=7, width=2, color='blue')
ax.tick_params(which='minor', length=4, width=1, color='red')

# Add major and minor gridlines
ax.grid(which='major', linestyle='-', linewidth='0.5', color='gray')
ax.grid(which='minor', linestyle=':', linewidth='0.25', color='lightgray')

plt.show()
```
