```python
import matplotlib.pyplot as plt
```

```python
year_of_exp = [2, 4, 6, 8, 10, 12, 14, 18, 20, 25]
median_salary = [
    45000,
    52000,
    60000,
    68000,
    75000,
    82000,
    90000,
    105000,
    115000,
    130000,
]
```

```python
plt.plot(
    year_of_exp,
    median_salary,
    marker="o",
    color="g",
    linestyle="-.",
    linewidth=1,
    markersize=4,
)

plt.xlabel("Years of Experience")
plt.ylabel("Median Salary ($)")
plt.title("Median Salary vs Years of Experience")
plt.grid(False)

plt.legend(["Median Salary"], loc="upper left")

plt.xticks(year_of_exp, fontsize=10, color="black")
# plt.xticks([]) # Hide the x-axis ticks
plt.yticks(median_salary, fontsize=10, color="red")

plt.savefig('Salary_VS_Experience.png', dpi=300, bbox_inches='tight')
plt.show()
```
