```python
import matplotlib.pyplot as plt
```

```python
# Subject names and scores
subjects = ['Math', 'Physics', 'Chemistry', 'English', 'CS']
scores = [85, 78, 91, 88, 95]
```

```python

# Plotting
plt.bar(subjects, scores)

# Labels and title
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Semester 1: Subject-wise Scores')

# Show plot
plt.show()
```
