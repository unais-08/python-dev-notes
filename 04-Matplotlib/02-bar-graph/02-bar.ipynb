```python
import matplotlib.pyplot as plt
```

```python
subjects = ["Math", "Physics", "Chemistry", "English", "CS"]
scores = [85, 78, 91, 88, 95]
colors = ["skyblue", "salmon", "lightgreen", "orange", "violet"]
```

```python
plt.bar(subjects, scores, color=colors)

plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Semester 1: Colored Subject-wise Scores")

plt.grid(axis="y", alpha=0.6)  # horizontal grid lines

plt.show()
```

```python
plt.barh(subjects, scores, color=colors, edgecolor="black")

plt.xlabel("Scores")
plt.ylabel("Subjects")
plt.title("Horizontal Bar Graph of Scores")

plt.text(
    scores[0] + 2, subjects[0], str(scores[0]), fontsize=10, color="black", ha="center"
)
plt.text(
    scores[1] + 2, subjects[1], str(scores[1]), fontsize=10, color="black", ha="center"
)
plt.text(
    scores[2] + 2, subjects[2], str(scores[2]), fontsize=10, color="black", ha="center"
)
plt.text(
    scores[3] + 2, subjects[3], str(scores[3]), fontsize=10, color="black", ha="center"
)
plt.text(
    scores[4] + 2, subjects[4], str(scores[4]), fontsize=10, color="black", ha="center"
)

plt.xlim(0, 105)  # Set x-axis limits
plt.grid(axis="x", alpha=0.5)  # vertical grid lines

plt.show()
```

```python
plt.bar(subjects, scores, color=colors, width=0.5)
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Scores with Labels")

# Add value labels on top of bars
for i in range(len(scores)):
    plt.text(i, scores[i] + 1, str(scores[i]), ha="center")

plt.ylim(0, 105)
plt.show()
```

```python
import numpy as np

subjects = ["Math", "Physics", "Chemistry", "English", "CS"]
student_a = [85, 78, 91, 88, 95]
student_b = [80, 75, 85, 82, 90]

x = np.arange(len(subjects))  # the label locations
print(x)
width = 0.35  # the width of the bars

plt.bar(x - width / 2, student_a, width, label="Student A", color="skyblue")
plt.bar(x + width / 2, student_b, width, label="Student B", color="lightgreen")

plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Score Comparison: Student A vs B")
plt.xticks(x, subjects)
plt.legend()
plt.grid(axis="y", alpha=0.6)  # horizontal grid lines
plt.show()
```
