```python
import matplotlib.pyplot as plt
```

```python
# Sample realistic data for 5 students
students = ["Aman", "Divya", "Ishaan", "Riya", "Kabir"]

ml_hours = [3, 5, 7, 9, 10]
ml_marks = [58, 68, 80, 92, 95]

cd_hours = [2, 4, 6, 8, 10]
cd_marks = [52, 60, 74, 86, 93]

math_hours = [1, 3, 5, 7, 9]
math_marks = [35, 48, 58, 70, 80]
```

```python
# Create figure
plt.figure(figsize=(10, 6))

# Plot each subject
plt.plot(ml_hours, ml_marks, marker="o", label="Machine Learning", linewidth=2)
plt.plot(cd_hours, cd_marks, marker="s", label="Compiler Design", linewidth=2)
plt.plot(math_hours, math_marks, marker="^", label="Mathematics III", linewidth=2)

# Add labels and title
plt.xlabel("Hours Studied", fontsize=12)
plt.ylabel("Marks Obtained", fontsize=12)
plt.title("Student Performance: Marks vs Study Hours", fontsize=14, fontweight='bold')

# Add grid, legend, limits
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Subjects", loc="lower right")
plt.xlim(0, 11)
plt.ylim(30, 100)

# Annotate student names for ML line
for i in range(len(students)):
    plt.annotate(students[i], (ml_hours[i], ml_marks[i]), textcoords="offset points", xytext=(0, 8), ha='center', fontsize=9)

# Annotate student names for CD line
# for i in range(len(students)):
#     plt.annotate(students[i], (cd_hours[i], cd_marks[i]), textcoords="offset points", xytext=(0, 8), ha='center', fontsize=9)

# Annotate student names for Math line
for i in range(len(students)):
    plt.annotate(students[i], (math_hours[i], math_marks[i]), textcoords="offset points", xytext=(0, 8), ha='center', fontsize=9)

# Show plot
plt.tight_layout()
plt.show()
```
