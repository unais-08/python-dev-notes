```python
plt.hist(marks, bins=10, edgecolor="black")

plt.title("Student Marks Distribution")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")

plt.grid(axis="y", alpha=0.5)
plt.show()
```

```python
# More detailed binning (smaller intervals)
plt.hist(marks, bins=20, edgecolor="black")

plt.title("Student Marks Distribution (More Bins)")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")

plt.grid(axis="y", alpha=0.5)
plt.show()
```

```python
# Define custom intervals (bins)
bins = [30, 40, 50, 60, 70, 80, 90, 100]

plt.hist(marks, bins=bins, edgecolor="black")

plt.title("Student Marks Distribution (Custom Ranges)")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")

plt.grid(axis="y", alpha=0.5)
plt.show()
```

```python
#  5. Histogram with Density Plot (Probability Distribution)
plt.hist(marks, bins=10, edgecolor="black", density=True)

plt.title("Normalized Histogram (Density)")
plt.xlabel("Marks Range")
plt.ylabel("Probability Density")

plt.grid(axis="y", alpha=0.5)
plt.show()
```

```python
#  6. Colored Histogram with Labels
plt.hist(marks, bins=10, color=["g"], edgecolor="black")

plt.title("Marks Distribution (Colored)")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")

plt.grid(axis="y", alpha=0.5)
plt.show()
```
