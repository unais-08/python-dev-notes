```python
import matplotlib.pyplot as plt
```

```python
x= [12, 14, 16, 18, 20]
y= [1000, 1500, 2000, 2500, 3000]
```

```python
# OOP style (Explicitly creating a figure and axes)
fig, ax = plt.subplots()  # Creates a figure and one Axes inside it

ax.plot(x,y)
ax.set_title('My First Plot')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')

# ax.set_xlim(0, 10)   # Axis limits
ax.xaxis.set_label_text("X Axis")  # Axis label
ax.xaxis.set_tick_params(rotation=45)  # Customize ticks
ax.yaxis.set_label_text("Y Axis")  # Axis label
plt.show()
```
