```python
import matplotlib.pyplot as plt
```

```python
subjects = ["Math", "Physics", "Chemistry", "English", "CS"]
time_spent = [5, 3, 4, 2, 6]  # in hours
```

```python
plt.pie(time_spent, labels=subjects)
plt.title("Time Spent on Subjects (in hours)")
plt.show()
```

```python
# Pie Chart with Percentages
plt.pie(time_spent, labels=subjects, autopct="%1.1f%%")
plt.title("Time Spent on Subjects (Percentage)")
plt.show()
```

```python
#  Pie Chart with Exploded Slice
explode = [0, 0, 0, 0, 0.1]  # explode only CS slice

plt.pie(time_spent, labels=subjects, autopct="%1.1f%%", explode=explode)

plt.title("Highlighting CS Subject")
plt.show()
```

```python
#  Colored Pie Chart
colors = ["skyblue", "lightgreen", "lightcoral", "gold", "violet"]

plt.pie(time_spent, labels=subjects, autopct="%1.1f%%", colors=colors)
plt.title("Time Spent on Subjects (Colored)")
plt.show()
```

```python
plt.pie(time_spent, labels=subjects, autopct="%1.1f%%", shadow=True, startangle=90)
plt.title("Pie Chart with Shadow and Start Angle")
# plt.text(0, -1.2, "Press the close button or ESC to exit the plot", ha='center', fontsize=10, color='gray')

# Place legend outside (right side)
plt.legend(title="Subjects", loc="center left", bbox_to_anchor=(1, 0.5))
plt.tight_layout()  # Auto adjust layout to prevent clipping
plt.show()
```
